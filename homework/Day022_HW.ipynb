### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁
    
'''
Your Code
'''
#find how many pages
import re
pages = soup.find_all('li',attrs={'class',re.compile('rc-pagination-item rc-pagination-item')})[-1]['title']
print(pages)
```

```python
# 2. 每頁都取出資料放入 courses
courses = []
#for i in '''Your Code''':
    
#    '''
#    Your Code
#    '''
#    browser.close()
print(pages)
i=0

for i in range (0,int(pages),1):
    print(i)
    basic_url = 'https://hahow.in/courses'
    insert_url = '?pages='+str(int(i)+1)
    if i==0 : 
        url=basic_url
    else :
        url = basic_url+insert_url
    browser = webdriver.Chrome(executable_path='../chromedriver')
    browser.get(url)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    for d in soup.find_all('div',attrs={'class':'sc-10r5mg2-0 fVNHJD hh-course-brief relative block'}):
        #print(1)
        title = d.find('h4').text
        description = d.find('div',attrs={'class':'meta-description'}).text
        courses.append({
            '描述': description,
            '標題': title
        })  
        #print(title)
        
    browser.close()
```

```python
# 3. done!

#print(courses)
import pandas as pd
pd.DataFrame(courses).reindex(columns=['標題','描述'])
```
