## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
import requests
from bs4 import BeautifulSoup
import random

proxy_ips = []

'''
Your Code
'''
url = 'https://free-proxy-list.net/'
resp = requests.get(url)
soup = BeautifulSoup(resp.text,'html5lib')
data_body = soup.find('tbody')
#data_body
for d in data_body.find_all('tr'):
    if d.find('td',attrs={'class':'hx'}).text=='yes':
        addr = d.find_all('td')[0].text
        port = d.find_all('td')[1].text
        proxy = addr + ':' + port
        proxy_ips.append(proxy)

print(proxy_ips)
```

```python
clean_ip = []
for i in range (0,len(proxy_ips)+1,1):
    #ip = random.choice(proxy_ips)
    ip = proxy_ips[i]
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=5)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip())
        clean_ip.append(ip)
    except:
        print('Fail')
proxy_ips = clean_ip
```
