### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_='part_list_2').find_all('h3'):
    print(d.find(class_='date').text , d.find('a').text)
    #print(d.get_text())
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# '''
#Your Code
#'''
class_list = []
for d in soup.find(class_='part_list_2').find_all('h3'):
    #print(d.find(class_='tag').text in class_list)
    if (d.find(class_='tag').text in class_list) == False:
        class_list.append(d.find(class_='tag').text)
#print(class_list)
      
data_list = []
count_list = []

for d in soup.find(class_='part_list_2').find_all('h3'):
    place = class_list.index(d.find(class_='tag').text)+1
    data = d.find(class_='date').text , d.find('a').text
    data_list.insert(place, data)
    count_list.append(d.find(class_='tag').text)
                          
for item in data_list:
    print(item)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
d = max(count_list,key=count_list.count)
print(d)
```
