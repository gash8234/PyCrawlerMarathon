## 1. 取出今天所有的新聞

```python
'''
Your Code
'''

import requests
import time
import re
from bs4 import BeautifulSoup
from selenium import webdriver

url = "https://www.ettoday.net/news/news-list.htm"

browser = webdriver.Chrome(executable_path='../chromedriver')

browser.get(url)
```

```python
##find out date today
html_source = browser.page_source
soup = BeautifulSoup(html_source,'html5lib')
date=soup.find(class_="part_list_2").find('h3').find(attrs={'class':'date'}).text
current_date=date.split(' ')[0].replace('/','.')

while 1:
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    html_source = browser.page_source
    
    ##check if current news belong to today or not 
    soup = BeautifulSoup(html_source,'html5lib')
    s=soup.find(class_="part_list_2").find_all('h3')[-1]
    comp = s.find(attrs={'class':'date'}).text
    if re.match(current_date,comp)==None : break
    print('today')
```

```python
soup = BeautifulSoup(html_source,'html5lib')
for d in soup.find(class_="part_list_2").find_all('h3') :
    date=d.find(attrs={'class':'date'}).text
    if re.match(current_date,date)!=None:
        print(date ,d.find('a',attrs={'target':'_blank'}).text)
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
soup = BeautifulSoup(html_source,'html5lib')
date=soup.find(class_="part_list_2").find('h3').find(attrs={'class':'date'}).text
current_date=date.split(' ')[0]
m2_hour=int(date.split(' ')[1][0:2])-2


for d in soup.find(class_="part_list_2").find_all('h3') :
    date=d.find(attrs={'class':'date'}).text
    hour=date.split(' ')[1][0:2]
    if re.match(current_date,date)!=None:
        if int(hour)>m2_hour:
            print(date ,d.find('a',attrs={'target':'_blank'}).text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''

##find out todays date
url = "https://www.ettoday.net/news/news-list.htm"
browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get(url)
html_source = browser.page_source
soup = BeautifulSoup(html_source,'html5lib')
date=soup.find(class_="part_list_2").find('h3').find(attrs={'class':'date'}).text
current_date=date.split(' ')[0].replace('/','-')
current_date=current_date[0:-2]+str(int(current_date[-2:])-3).zfill(2)
print(current_date)
browser.close()

#url = "https://www.ettoday.net/news/news-list"+'-'+current_date.replace('/','-')+'-0.htm'
```

```python
url = "https://www.ettoday.net/news/news-list"+'-'+current_date.replace('/','-')+'-0.htm'
print(url)

browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get(url)
```

```python
while 1:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    html_source = browser.page_source
    
    ##check if current news belong to today or not 
    soup = BeautifulSoup(html_source,'html5lib')
    s=soup.find(class_="part_list_2").find_all('h3')[-1]
    comp = s.find(attrs={'class':'date'}).text.split(' ')[1][0:2]
    if int(comp)<15 : break
    print(comp)
```

```python
for d in soup.find(class_="part_list_2").find_all('h3') :
    date=d.find(attrs={'class':'date'}).text
    hour=date.split(' ')[1][0:2]
    if int(hour)>=15 and int(hour)<=17:
        print(date ,d.find('a',attrs={'target':'_blank'}).text)

browser.close()
```
