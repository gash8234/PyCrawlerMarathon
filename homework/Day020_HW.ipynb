```python
page_key='24d59309-4d23-460e-b5c3-904b91e284f1'
while 1:
    url='https://www.dcard.tw/service/api/v2/popularForums/GetPage?pageKey='+page_key
    r=requests.get(url)
    d=json.loads(r.text)
    for data in d['items']:
        print(data['name'])
    page_key=d['nextKey']
    if d['nextKey']=='': break
    
```
