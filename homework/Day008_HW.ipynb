### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

'''
Your Code Here
'''
headers = {'user-agent':'user'}
r = requests.get(url , headers = headers)

r.encoding = 'utf-8'
print(r.text[0:1000])
print()
soup = BeautifulSoup(r.text,features = 'html.parser')
print(soup)
```
